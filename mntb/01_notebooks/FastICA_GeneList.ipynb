```python
# Alphabetical order is standard
# We're doing "import superlongname as abbrev" for our laziness - this way we don't have to type out the whole thing each time.

# Python plotting library
import matplotlib.pyplot as plt

# Numerical python library (pronounced "num-pie")
import numpy as np

# Dataframes in Python
import pandas as pd

# Statistical plotting library we'll use
import seaborn as sns

# This is necessary to show the plotted figures inside the notebook -- "inline" with the notebook cells
%matplotlib inline
```

```python
counts = pd.read_csv('C:/Users/anbra/Desktop/Single_Cell_MNTB/Data/FinalsCountsMatrix_GoodGenes.csv', sep ='\t')
print(counts.shape)
counts.head()
```

```python
counts = counts.set_index('GeneID')
counts.head()
```

```python
from sklearn.decomposition import PCA, FastICA, NMF
from sklearn.manifold import TSNE, MDS
```

```python
counts_t = counts.T
print(counts_t.shape)
counts_t.head()
```

```python
smusher = FastICA(n_components=40)

smushed = pd.DataFrame(smusher.fit_transform(counts_t), index=counts_t.index)
smushed.head()
```

```python
smushed.columns.name = 'ICA Components'
```

```python
components = pd.DataFrame(smusher.components_, columns=counts_t.columns).T
print(components.shape)
components.head()
```

```python
component_sizes = components.apply(np.linalg.norm)

plt.plot(component_sizes, 'o')
```

```python
component_ranks = component_sizes.rank().sort_values(ascending=False)
component_ranks
```

```python
plt.plot(component_sizes[component_ranks.index].values, 'o')
```

```python
is_positive = components.max() == components.abs().max()
is_positive
```

```python
multiplier = is_positive.astype(int)
multiplier[~is_positive] = -1
multiplier
```

```python
components_flipped = components * multiplier
print(components_flipped.shape)
components_flipped.head()
```

```python
component_ranks = components_flipped.rank(ascending=False)
component_ranks.head()
```

```python
top_genes = component_ranks[component_ranks < 25]
top_genes.head()
```

```python
top_genes_dropna = top_genes.dropna(how='all')
print(top_genes_dropna.shape)
top_genes_dropna.head()
```

```python
top_genes_dropna
```

```python
ICA_genes = counts.loc[top_genes_dropna.index]
```
