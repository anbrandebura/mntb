```python
# Alphabetical order is standard
# We're doing "import superlongname as abbrev" for our laziness - this way we don't have to type out the whole thing each time.

# Python plotting library
import matplotlib.pyplot as plt

# Numerical python library (pronounced "num-pie")
import numpy as np

# Dataframes in Python
import pandas as pd

# Statistical plotting library we'll use
import seaborn as sns

# This is necessary to show the plotted figures inside the notebook -- "inline" with the notebook cells
%matplotlib inline
```

```python
#Read in counts matrix
countsMatrix = pd.read_csv('C:/Users/anbra/Desktop/Single_Cell_MNTB/Data/CellFilter_EndogGenes_TCsRemoved.csv', sep = '\t')
countsMatrix.head()
```

```python
#Make GeneID the index
countsMatrix_2 = countsMatrix.set_index(['GeneID'])
print(countsMatrix_2.shape)
countsMatrix_2.head()
```

```python
#plot TPPM
sns.boxplot(countsMatrix_2)

# gcf = Get current figure
fig = plt.gcf()
```

```python
#Find genes with greater than or equal to 1 TPM expression level in at least 3 cells
genes_of_interest = (countsMatrix_2 >= 1).sum(axis=1) >= 3
genes_of_interest.head()
```

```python
#Subset the matrix to get only the genes of interest
expression_filtered = countsMatrix_2.loc[genes_of_interest]
print(expression_filtered.shape)  
expression_filtered.head()
```

```python
#Find CV for these genes 
cv = np.std(expression_filtered, axis = 1)/ np.mean(expression_filtered, axis = 1)
print(cv.shape)
cv.head()
```

```python
#Plot CV of genes
sns.boxplot(cv)

# gcf = Get current figure
fig = plt.gcf()
```

```python
#Find genes withCV of greater than or equal to 2.5
filtered_cv_genes = cv >= 2.5
filtered_cv_genes.sum()
```

```python
#Subset the matrix on these genes
expression_filtered_2 = expression_filtered.loc[filtered_cv_genes]
print(expression_filtered_2.shape)
expression_filtered_2.head()
```

```python
#Save new matrix with uninteresting genes removed
expression_filtered_2.to_csv('C:/Users/anbra/Desktop/Single_Cell_MNTB/Data/FinalsCountsMatrix_GoodGenes.csv', sep = '\t')
```
